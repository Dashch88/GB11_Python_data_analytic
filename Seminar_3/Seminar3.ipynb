```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin-1')
df.head()
```

### Задание 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df['Cpu_Company'].value_counts()
```

### Задание 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя

```python
def mem_amount(memory_str: str):
    memory_str = memory_str.split()[0]
    if memory_str.endswith('GB'):
        return memory_str.replace('GB', '')
    elif memory_str.endswith('TB'):
        return int(float(memory_str.replace('TB', ''))) * 1024

df['Memory_Amount'] = df.Memory.apply(mem_amount)
df.Memory_Amount.value_counts()
```

### Задание 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

```python
def mem_type_amount(memory_str: str):
    if memory_str.split()[1] == 'Flash':
        return 'Flash Storage'
    else:
        return memory_str.split()[1]

mem_type_amount('128GB Flash Storage')

df['Memory_Type'] = df.Memory.apply(mem_type_amount)
df.Memory_Type.value_counts()
```

### Задание 1.4 Удалите признаки Memory и ScreenResolution

```python
df_new = df.drop(columns=['Memory', 'ScreenResolution'])
```
